```python
import pywt
import scaleogram as scg 
import numpy as np
from skimage.transform import resize
from tqdm import tqdm
```

```python
wavelet='mexh'
signal_length=64
# wavelet = 'morl' # mother wavelet
scales =scg.periods2scales( np.arange(1, signal_length+1) ) # range of scales 

train_scalo_df=[]
for i in tqdm(range(len(train))):
    coeffs, freqs= pywt.cwt(train.iloc[i,4:], scales,wavelet=wavelet) 
    rescale_coeffs = resize((coeffs), (signal_length, signal_length), mode = 'constant')

    train_scalo_df.append(rescale_coeffs)
    
```

```python
valid_scalo_df=[]
for i in tqdm(range(len(valid))):
    coeffs, freqs= pywt.cwt(valid.iloc[i,4:], scales,wavelet=wavelet) 
    rescale_coeffs = resize((coeffs), (signal_length, signal_length), mode = 'constant')

    valid_scalo_df.append(rescale_coeffs)
```

```python
test_scalo_df=[]
for i in tqdm(range(len(test))):
    coeffs, freqs= pywt.cwt(test.iloc[i,4:], scales,wavelet=wavelet) 
    rescale_coeffs = resize((coeffs), (signal_length, signal_length), mode = 'constant')

    test_scalo_df.append(rescale_coeffs)
```

```python
###
```

```python
train_scalo_df=np.array(train_scalo_df)
valid_scalo_df=np.array(valid_scalo_df)
test_scalo_df=np.array(test_scalo_df)
```
